```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
t=2*np.pi
l=1
r=1
c=1
vo=1
w=np.arange(-8*np.pi,8*np.pi,.001)

z=(((l/c)**2+((r/(w*c)-r*w*l)**2))**(1/2))/(l/(r*c)+c/(l*r)*(r*w*l-r/(w*c))**2)
phi=np.arctan(r/(w*l)-r*w*c)
I=(vo*np.cos(w*t-phi))/z
plt.plot(w,I,'r')
plt.xlabel('Frequency')
plt.ylabel('Current')
plt.title('Current vs Frequency')
```
