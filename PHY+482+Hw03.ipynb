```python
#worked with erick verleyee
import matplotlib.pyplot as plt
import numpy as np
% matplotlib inline
```

```python


x0 = 3
w = 8
v = 10000
B0 = .2
t=0
dt=.01
f=[]
t_1=[]
for i in range(500):
    EMF = (10000*w*(x0+v*t))+(B0+10000*t)*w*v
    f.append(EMF)
    t_1.append(t)
    t+=dt
plt.plot(t_1,f)
plt.xlabel('time')
plt.ylabel('EMF')
plt.title('EMF vs Time')
```
