```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

w=np.arange(0,50,.1)

R=1
L=1
V=R/(w**2*L**2+R**2)**(1/2)

plt.plot(w,V,'r')
plt.xlabel('Frequency')
plt.ylabel('Vout/Vin')
```

This describes a low pass filter. This is evident by the decaying nature of the curve.

```python
w=1
v1=[]
t1=[]
dt=.01
t=0

while t<100:
    V=[]
    for k in range (1,300):
        v=4/np.pi*np.sin((2*k-1)*w*t)/(2*k-1)
        V.append(v)
        
    v1.append(sum(V))
    t1.append(t)
    t+=dt
    
plt.plot(t1,v1)
plt.xlabel('time')
plt.ylabel('vin')
    
    



        
    
```
